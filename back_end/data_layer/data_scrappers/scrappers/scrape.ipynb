```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# List of schedule URLs
SCHEDULE_URLS = [
    'https://www.basketball-reference.com/leagues/NBA_2020_games-october-2019.html',
    'https://www.basketball-reference.com/leagues/NBA_2020_games-november.html',
    'https://www.basketball-reference.com/leagues/NBA_2020_games-december.html',
    'https://www.basketball-reference.com/leagues/NBA_2020_games-january.html',
    'https://www.basketball-reference.com/leagues/NBA_2020_games-february.html',
    'https://www.basketball-reference.com/leagues/NBA_2020_games-march.html',
    'https://www.basketball-reference.com/leagues/NBA_2020_games-july.html',
    'https://www.basketball-reference.com/leagues/NBA_2020_games-august.html',
    'https://www.basketball-reference.com/leagues/NBA_2020_games-september.html',
    'https://www.basketball-reference.com/leagues/NBA_2020_games-october-2020.html',
    'https://www.basketball-reference.com/leagues/NBA_2021_games-december.html',
    'https://www.basketball-reference.com/leagues/NBA_2021_games-january.html',
    'https://www.basketball-reference.com/leagues/NBA_2021_games-february.html',
    'https://www.basketball-reference.com/leagues/NBA_2021_games-march.html',
    'https://www.basketball-reference.com/leagues/NBA_2021_games-april.html',
    'https://www.basketball-reference.com/leagues/NBA_2021_games-may.html',
    'https://www.basketball-reference.com/leagues/NBA_2021_games-june.html',
    'https://www.basketball-reference.com/leagues/NBA_2021_games-july.html',
    'https://www.basketball-reference.com/leagues/NBA_2022_games-october.html',
    'https://www.basketball-reference.com/leagues/NBA_2022_games-november.html',
    'https://www.basketball-reference.com/leagues/NBA_2022_games-december.html',
    'https://www.basketball-reference.com/leagues/NBA_2022_games-january.html',
    'https://www.basketball-reference.com/leagues/NBA_2022_games-february.html',
    'https://www.basketball-reference.com/leagues/NBA_2022_games-march.html',
    'https://www.basketball-reference.com/leagues/NBA_2022_games-april.html',
    'https://www.basketball-reference.com/leagues/NBA_2022_games-may.html',
    'https://www.basketball-reference.com/leagues/NBA_2022_games-june.html',
    'https://www.basketball-reference.com/leagues/NBA_2023_games-october.html',
    'https://www.basketball-reference.com/leagues/NBA_2023_games-november.html',
    'https://www.basketball-reference.com/leagues/NBA_2023_games-december.html',
    'https://www.basketball-reference.com/leagues/NBA_2023_games-january.html',
    'https://www.basketball-reference.com/leagues/NBA_2023_games-february.html',
    'https://www.basketball-reference.com/leagues/NBA_2023_games-march.html',
    'https://www.basketball-reference.com/leagues/NBA_2023_games-april.html',
    'https://www.basketball-reference.com/leagues/NBA_2023_games-may.html',
    'https://www.basketball-reference.com/leagues/NBA_2023_games-june.html',
    'https://www.basketball-reference.com/leagues/NBA_2024_games-october.html',
    'https://www.basketball-reference.com/leagues/NBA_2024_games-november.html',
    'https://www.basketball-reference.com/leagues/NBA_2024_games-december.html',
    'https://www.basketball-reference.com/leagues/NBA_2024_games-january.html',
    'https://www.basketball-reference.com/leagues/NBA_2024_games-february.html',
    'https://www.basketball-reference.com/leagues/NBA_2024_games-march.html',
    'https://www.basketball-reference.com/leagues/NBA_2024_games-april.html',
    'https://www.basketball-reference.com/leagues/NBA_2024_games-may.html',
    'https://www.basketball-reference.com/leagues/NBA_2024_games-june.html',
    'https://www.basketball-reference.com/leagues/NBA_2025_games-october.html',
    'https://www.basketball-reference.com/leagues/NBA_2025_games-november.html',
    'https://www.basketball-reference.com/leagues/NBA_2025_games-december.html',
    'https://www.basketball-reference.com/leagues/NBA_2025_games-january.html',
    'https://www.basketball-reference.com/leagues/NBA_2025_games-february.html',
    'https://www.basketball-reference.com/leagues/NBA_2025_games-march.html',
    'https://www.basketball-reference.com/leagues/NBA_2025_games-april.html',
    'https://www.basketball-reference.com/leagues/NBA_2025_games-may.html',
    'https://www.basketball-reference.com/leagues/NBA_2025_games-june.html'
]

def setup_driver():
    """Setup Chrome driver with options"""
    chrome_options = Options()
    chrome_options.add_argument('--headless')  # Run in background
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    chrome_options.add_argument('--disable-gpu')
    chrome_options.add_argument('--window-size=1920,1080')
    
    # You might need to specify the path to your chromedriver
    # service = Service('/path/to/chromedriver')
    # driver = webdriver.Chrome(service=service, options=chrome_options)
    
    driver = webdriver.Chrome(options=chrome_options)
    driver.implicitly_wait(10)
    return driver

def get_box_score_links(driver, url):
    """Get all box score links from a schedule page"""
    box_score_links = []
    
    try:
        print(f"Processing: {url}")
        driver.get(url)
        
        # Wait for the table to load
        wait = WebDriverWait(driver, 10)
        table = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "table#schedule")))
        
        # Find all box score links in the table
        # Box score links are typically in cells with data-stat="box_score_text"
        box_score_cells = table.find_elements(By.CSS_SELECTOR, "td[data-stat='box_score_text'] a")
        
        for cell in box_score_cells:
            link = cell.get_attribute('href')
            if link and 'boxscores' in link:
                box_score_links.append(link)
        
        print(f"Found {len(box_score_links)} box score links")
        
    except Exception as e:
        print(f"Error processing {url}: {e}")
    
    return box_score_links

def scrape_all_box_scores():
    """Scrape box score links from all schedule URLs"""
    all_box_score_links = []
    driver = setup_driver()
    
    try:
        for url in SCHEDULE_URLS:
            links = get_box_score_links(driver, url)
            all_box_score_links.extend(links)
            time.sleep(2)  # Be respectful to the server
            
        print(f"\nTotal box score links collected: {len(all_box_score_links)}")
        
        # Print first few links as sample
        if all_box_score_links:
            print("\nSample box score links:")
            for i, link in enumerate(all_box_score_links[:5]):
                print(f"{i+1}. {link}")
                
    finally:
        driver.quit()
    
    return all_box_score_links

def save_links_to_file(links, filename="box_score_links.txt"):
    """Save links to a text file"""
    with open(filename, 'w') as f:
        for link in links:
            f.write(f"{link}\n")
    print(f"\nLinks saved to {filename}")

if __name__ == "__main__":
    # Scrape all box score links
    box_score_links = scrape_all_box_scores()
    
    # Save to file
    save_links_to_file(box_score_links)
    
    # Also print summary
    print(f"\nSummary:")
    print(f"Total box score links: {len(box_score_links)}")
    
    # Count by season (optional)
    seasons = {}
    for link in box_score_links:
        if '/boxscores/' in link:
            parts = link.split('/')
            if len(parts) > 4:
                year_month = parts[4]
                if year_month not in seasons:
                    seasons[year_month] = 0
                seasons[year_month] += 1
    
    print("\nLinks by season month:")
    for season, count in sorted(seasons.items()):
        print(f"{season}: {count} games")
```

```python
# Function to extract game info from URL
def extract_game_info(url):
    """
    Extract game date, home team, away team from URL
    URL format: https://www.basketball-reference.com/boxscores/202310240LAL.html
    """
    # Extract game ID from URL
    game_id = url.split('/')[-1].replace('.html', '')
    
    if len(game_id) >= 11:
        # First 8 characters are date (YYYYMMDD)
        date_str = game_id[:8]
        try:
            game_date = datetime.strptime(date_str, '%Y%m%d').strftime('%Y-%m-%d')
        except:
            game_date = date_str
        
        # Next characters are home team, last 3 are away team
        # Example: 202310240LAL -> home: 0 (means unspecified?), away: LAL
        # Actually format is: YYYYMMDD0XXX where XXX is home team
        
        if len(game_id) == 12:  # Standard format
            home_team = game_id[8:-3]
            away_team = game_id[-3:]
        else:
            # Try to extract teams from the page instead
            home_team = "Unknown"
            away_team = "Unknown"
        
        return {
            'game_id': game_id,
            'game_date': game_date,
            'url': url
        }
    return None

# Test extraction
game_info = extract_game_info('https://www.basketball-reference.com/boxscores/201910230ORL.html')
game_info
```

```python
# 2. TABLE-SPECIFIC PARSING FUNCTIONS

def parse_line_score_table(soup, url):
    """Parse the line score table (quarter-by-quarter scoring)"""
    table = soup.find('table', {'id': 'line_score'})
    if not table:
        return None
    
    # Extract game info from URL
    game_id = url.split('/')[-1].replace('.html', '')
    if len(game_id) >= 8:
        game_date = f"{game_id[:4]}-{game_id[4:6]}-{game_id[6:8]}"
    else:
        game_date = "Unknown"
    
    rows = []
    # Find team rows
    tbody = table.find('tbody')
    if not tbody:
        return None
    
    for row in tbody.find_all('tr'):
        team_cell = row.find('th', {'data-stat': 'team'})
        if not team_cell:
            continue
        
        team_name = team_cell.text.strip()
        if not team_name:
            team_link = team_cell.find('a')
            if team_link:
                team_name = team_link.text.strip()
        
        row_data = {
            'game_date': game_date,
            'game_id': game_id,
            'team': team_name,
            'url': url
        }
        
        # Get all quarter scores
        for quarter in ['1', '2', '3', '4']:
            q_cell = row.find('td', {'data-stat': quarter})
            if q_cell:
                row_data[f'Q{quarter}'] = q_cell.text.strip()
        
        # Get OT scores
        for ot_num in range(1, 6):  # Support up to 5 OTs
            ot_cell = row.find('td', {'data-stat': f'{ot_num}OT'})
            if ot_cell:
                row_data[f'OT{ot_num}'] = ot_cell.text.strip()
        
        # Get total
        total_cell = row.find('td', {'data-stat': 'T'})
        if total_cell:
            row_data['TOTAL'] = total_cell.text.strip()
        
        rows.append(row_data)
    
    return rows

def parse_four_factors_table(soup, url):
    """Parse the Four Factors table (team advanced stats)"""
    table = soup.find('table', {'id': 'four_factors'})
    if not table:
        return None
    
    game_id = url.split('/')[-1].replace('.html', '')
    if len(game_id) >= 8:
        game_date = f"{game_id[:4]}-{game_id[4:6]}-{game_id[6:8]}"
    else:
        game_date = "Unknown"
    
    rows = []
    tbody = table.find('tbody')
    if not tbody:
        return None
    
    for row in tbody.find_all('tr'):
        team_cell = row.find('th', {'data-stat': 'team_id'})
        if not team_cell:
            continue
        
        team_name = team_cell.text.strip()
        if not team_name:
            team_link = team_cell.find('a')
            if team_link:
                team_name = team_link.text.strip()
        
        row_data = {
            'game_date': game_date,
            'game_id': game_id,
            'team': team_name,
            'url': url
        }
        
        # Four factors stats
        stats_mapping = {
            'pace': 'pace',
            'efg_pct': 'eFG%',
            'tov_pct': 'TOV%',
            'orb_pct': 'ORB%',
            'ft_rate': 'FT/FGA',
            'off_rtg': 'ORtg'
        }
        
        for stat_id, stat_name in stats_mapping.items():
            cell = row.find('td', {'data-stat': stat_id})
            if cell:
                value = cell.text.strip()
                # Clean percentage values
                if stat_id in ['efg_pct', 'tov_pct', 'orb_pct', 'ft_rate']:
                    if value.startswith('.'):
                        try:
                            value = float(value)
                        except:
                            pass
                row_data[stat_name] = value
        
        rows.append(row_data)
    
    return rows

def parse_team_basic_table(soup, url, table_id):
    """Parse team basic box score table"""
    table = soup.find('table', {'id': table_id})
    if not table:
        return None
    
    # Extract team from table ID (e.g., 'box-HOU-game-basic' -> 'HOU')
    team_match = re.search(r'box-(\w+)-', table_id)
    team = team_match.group(1) if team_match else "Unknown"
    
    game_id = url.split('/')[-1].replace('.html', '')
    if len(game_id) >= 8:
        game_date = f"{game_id[:4]}-{game_id[4:6]}-{game_id[6:8]}"
    else:
        game_date = "Unknown"
    
    rows = []
    tbody = table.find('tbody')
    if not tbody:
        return None
    
    for row in tbody.find_all('tr'):
        # Skip header rows
        if row.get('class') and any(x in row.get('class') for x in ['thead', 'over_header']):
            continue
        
        player_cell = row.find('th', {'data-stat': 'player'})
        if not player_cell:
            continue
        
        player_name = player_cell.text.strip()
        
        # Skip "Did Not Play" rows
        reason_cell = row.find('td', {'data-stat': 'reason'})
        if reason_cell and reason_cell.text.strip():
            continue
        
        # Skip team totals (we'll handle separately)
        if 'Team Totals' in player_name:
            continue
        
        row_data = {
            'game_date': game_date,
            'game_id': game_id,
            'team': team,
            'player': player_name.replace('*', '').strip(),
            'url': url
        }
        
        # Basic stats to collect
        basic_stats = [
            'mp', 'fg', 'fga', 'fg_pct', 'fg3', 'fg3a', 'fg3_pct',
            'ft', 'fta', 'ft_pct', 'orb', 'drb', 'trb', 'ast',
            'stl', 'blk', 'tov', 'pf', 'pts', 'game_score', 'plus_minus'
        ]
        
        for stat in basic_stats:
            cell = row.find('td', {'data-stat': stat})
            if cell:
                value = cell.text.strip()
                if value and value not in ['', 'Did Not Play', 'DNP']:
                    # Convert minutes from MM:SS to decimal
                    if stat == 'mp' and ':' in value:
                        try:
                            mins, secs = map(int, value.split(':'))
                            row_data[stat] = round(mins + secs/60, 2)
                        except:
                            row_data[stat] = value
                    else:
                        row_data[stat] = value
        
        # Only add if we have actual stats
        if len(row_data) > 5:
            rows.append(row_data)
    
    return rows

def parse_team_advanced_table(soup, url, table_id):
    """Parse team advanced box score table"""
    table = soup.find('table', {'id': table_id})
    if not table:
        return None
    
    team_match = re.search(r'box-(\w+)-', table_id)
    team = team_match.group(1) if team_match else "Unknown"
    
    game_id = url.split('/')[-1].replace('.html', '')
    if len(game_id) >= 8:
        game_date = f"{game_id[:4]}-{game_id[4:6]}-{game_id[6:8]}"
    else:
        game_date = "Unknown"
    
    rows = []
    tbody = table.find('tbody')
    if not tbody:
        return None
    
    for row in tbody.find_all('tr'):
        if row.get('class') and any(x in row.get('class') for x in ['thead', 'over_header']):
            continue
        
        player_cell = row.find('th', {'data-stat': 'player'})
        if not player_cell:
            continue
        
        player_name = player_cell.text.strip()
        
        # Skip "Did Not Play" rows
        reason_cell = row.find('td', {'data-stat': 'reason'})
        if reason_cell and reason_cell.text.strip():
            continue
        
        if 'Team Totals' in player_name:
            continue
        
        row_data = {
            'game_date': game_date,
            'game_id': game_id,
            'team': team,
            'player': player_name.replace('*', '').strip(),
            'url': url
        }
        
        # Advanced stats
        advanced_stats = [
            'mp', 'ts_pct', 'efg_pct', 'fg3a_per_fga_pct', 'fta_per_fga_pct',
            'orb_pct', 'drb_pct', 'trb_pct', 'ast_pct', 'stl_pct', 'blk_pct',
            'tov_pct', 'usg_pct', 'off_rtg', 'def_rtg', 'bpm'
        ]
        
        for stat in advanced_stats:
            cell = row.find('td', {'data-stat': stat})
            if cell:
                value = cell.text.strip()
                if value and value not in ['', 'Did Not Play', 'DNP']:
                    row_data[stat] = value
        
        if len(row_data) > 5:
            rows.append(row_data)
    
    return rows

def parse_quarter_table(soup, url, table_id):
    """Parse quarter-specific tables (Q1, Q2, Q3, Q4, OT1, OT2, etc.)"""
    table = soup.find('table', {'id': table_id})
    if not table:
        return None
    
    # Extract team and quarter from table ID
    match = re.search(r'box-(\w+)-(\w+)-basic', table_id)
    if not match:
        return None
    
    team = match.group(1)
    quarter = match.group(2)  # q1, q2, q3, q4, ot1, ot2, etc
    
    game_id = url.split('/')[-1].replace('.html', '')
    if len(game_id) >= 8:
        game_date = f"{game_id[:4]}-{game_id[4:6]}-{game_id[6:8]}"
    else:
        game_date = "Unknown"
    
    rows = []
    tbody = table.find('tbody')
    if not tbody:
        return None
    
    for row in tbody.find_all('tr'):
        if row.get('class') and any(x in row.get('class') for x in ['thead', 'over_header']):
            continue
        
        player_cell = row.find('th', {'data-stat': 'player'})
        if not player_cell:
            continue
        
        player_name = player_cell.text.strip()
        
        # Skip "Did Not Play" rows
        reason_cell = row.find('td', {'data-stat': 'reason'})
        if reason_cell and reason_cell.text.strip():
            continue
        
        if 'Team Totals' in player_name:
            continue
        
        row_data = {
            'game_date': game_date,
            'game_id': game_id,
            'team': team,
            'quarter': quarter.upper(),
            'player': player_name.replace('*', '').strip(),
            'url': url
        }
        
        # Quarter stats
        quarter_stats = [
            'mp', 'fg', 'fga', 'fg_pct', 'fg3', 'fg3a', 'fg3_pct',
            'ft', 'fta', 'ft_pct', 'orb', 'drb', 'trb', 'ast',
            'stl', 'blk', 'tov', 'pf', 'pts', 'game_score', 'plus_minus'
        ]
        
        for stat in quarter_stats:
            cell = row.find('td', {'data-stat': stat})
            if cell:
                value = cell.text.strip()
                if value and value not in ['', 'Did Not Play', 'DNP']:
                    if stat == 'mp' and ':' in value:
                        try:
                            mins, secs = map(int, value.split(':'))
                            row_data[stat] = round(mins + secs/60, 2)
                        except:
                            row_data[stat] = value
                    else:
                        row_data[stat] = value
        
        if len(row_data) > 6:  # More than just metadata
            rows.append(row_data)
    
    return rows

print("‚úÖ All parsing functions ready")
```

```python
# 3. MAIN SCRAPING FUNCTION

def scrape_game_page(url, driver=None):
    """Scrape all tables from a game page"""
    print(f"\nüîç Scraping: {url}")
    
    close_driver = False
    if driver is None:
        driver = setup_driver(headless=True)
        close_driver = True
    
    try:
        driver.get(url)
        time.sleep(4)  # Wait for page to load
        
        # Check if page loaded successfully
        if "access denied" in driver.page_source.lower():
            print("‚ùå Access denied by website")
            return None
        
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        # Find all tables to identify what's available
        all_tables = soup.find_all('table')
        print(f"Found {len(all_tables)} tables total")
        
        # Initialize data storage
        all_data = {
            'line_scores': [],
            'four_factors': [],
            'team_basic': [],
            'team_advanced': [],
            'quarters': []
        }
        
        # Parse each table type
        print("\nüìä Parsing tables...")
        
        # 1. Line Score table
        line_scores = parse_line_score_table(soup, url)
        if line_scores:
            all_data['line_scores'] = line_scores
            print(f"  ‚úì Line scores: {len(line_scores)} rows")
        
        # 2. Four Factors table
        four_factors = parse_four_factors_table(soup, url)
        if four_factors:
            all_data['four_factors'] = four_factors
            print(f"  ‚úì Four factors: {len(four_factors)} rows")
        
        # 3. Find all team basic tables
        basic_tables = soup.find_all('table', id=re.compile(r'box-\w+-game-basic$'))
        for table in basic_tables:
            table_id = table.get('id')
            table_data = parse_team_basic_table(soup, url, table_id)
            if table_data:
                all_data['team_basic'].extend(table_data)
        
        if all_data['team_basic']:
            print(f"  ‚úì Team basic stats: {len(all_data['team_basic'])} rows")
        
        # 4. Find all team advanced tables
        advanced_tables = soup.find_all('table', id=re.compile(r'box-\w+-game-advanced$'))
        for table in advanced_tables:
            table_id = table.get('id')
            table_data = parse_team_advanced_table(soup, url, table_id)
            if table_data:
                all_data['team_advanced'].extend(table_data)
        
        if all_data['team_advanced']:
            print(f"  ‚úì Team advanced stats: {len(all_data['team_advanced'])} rows")
        
        # 5. Find quarter tables (Q1, Q2, Q3, Q4, OT1, OT2, etc.)
        quarter_tables = soup.find_all('table', id=re.compile(r'box-\w+-(q\d+|ot\d+)-basic$'))
        for table in quarter_tables:
            table_id = table.get('id')
            table_data = parse_quarter_table(soup, url, table_id)
            if table_data:
                all_data['quarters'].extend(table_data)
        
        if all_data['quarters']:
            print(f"  ‚úì Quarter stats: {len(all_data['quarters'])} rows")
        
        # Summary
        total_rows = sum(len(data) for data in all_data.values())
        print(f"\n‚úÖ Total rows collected: {total_rows}")
        
        return all_data
        
    except Exception as e:
        print(f"‚ùå Error scraping {url}: {str(e)}")
        import traceback
        traceback.print_exc()
        return None
        
    finally:
        if close_driver and driver:
            driver.quit()

print("‚úÖ Main scraping function ready")
```

```python
# 4. SAVE TO CSV FUNCTIONS

def save_to_csv(data_dict, game_url, output_dir="nba_boxscores"):
    """Save all data to separate CSV files"""
    
    # Create output directory
    os.makedirs(output_dir, exist_ok=True)
    
    # Extract game ID for filename
    game_id = game_url.split('/')[-1].replace('.html', '')
    
    saved_files = []
    
    # Save each data type to separate CSV
    for data_type, data_list in data_dict.items():
        if not data_list:
            continue
        
        df = pd.DataFrame(data_list)
        
        # Create filename
        filename = f"{output_dir}/{data_type}_{game_id}.csv"
        df.to_csv(filename, index=False, encoding='utf-8')
        
        saved_files.append((data_type, filename, len(df)))
        
        print(f"üíæ Saved {data_type}: {len(df)} rows ‚Üí {filename}")
    
    # Also create a summary file
    summary_file = f"{output_dir}/summary_{game_id}.txt"
    with open(summary_file, 'w') as f:
        f.write(f"Game: {game_url}\n")
        f.write(f"Scraped: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n\n")
        
        for data_type, filename, row_count in saved_files:
            f.write(f"{data_type}: {row_count} rows ({filename})\n")
    
    print(f"\nüìù Summary saved to: {summary_file}")
    
    return saved_files

print("‚úÖ CSV saving functions ready")
```

```python
# 5. LOAD LINKS AND RUN SCRAPER

def load_links(filename="box_score_links.txt"):
    """Load box score links from file"""
    try:
        with open(filename, 'r') as f:
            links = [line.strip() for line in f if line.strip()]
        print(f"üì• Loaded {len(links)} links from {filename}")
        return links
    except FileNotFoundError:
        print(f"‚ùå File {filename} not found!")
        return []

# Load links
box_score_links = load_links()

if not box_score_links:
    # Use a sample URL
    box_score_links = ['https://www.basketball-reference.com/boxscores/202310240LAL.html']
    print(f"üìù Using sample URL: {box_score_links[0]}")

print("‚úÖ Links loaded")
```

```python
# 7. BATCH PROCESSING FUNCTION

def batch_scrape_games(url_list, start_from=0, max_games=None, delay=3):
    """Scrape multiple games in batch"""
    
    total_games = len(url_list[start_from:start_from + max_games]) if max_games else len(url_list[start_from:])
    
    print(f"\nüì¶ BATCH PROCESSING")
    print(f"Games to scrape: {total_games}")
    print(f"Start from index: {start_from}")
    print("=" * 60)
    
    driver = setup_driver(headless=True)
    
    try:
        for i, url in enumerate(url_list[start_from:], start_from):
            if max_games and (i - start_from) >= max_games:
                print(f"\n‚úÖ Reached maximum games limit ({max_games})")
                break
            
            print(f"\n[{i+1}/{len(url_list)}] ", end="")
            game_data = scrape_game_page(url, driver)
            
            if game_data:
                save_to_csv(game_data, url)
            else:
                print(f"‚ö†Ô∏è Skipping game {i+1}")
            
            # Add delay between requests
            if i < len(url_list) - 1:
                print(f"‚è≥ Waiting {delay} seconds...")
                time.sleep(delay)
        
        print("\nüéâ BATCH PROCESSING COMPLETE!")
        
    except KeyboardInterrupt:
        print("\n‚ö†Ô∏è Batch processing interrupted by user")
    except Exception as e:
        print(f"\n‚ùå Batch processing error: {str(e)}")
    finally:
        driver.quit()
        print("‚úÖ Driver closed")

print("‚úÖ Batch processing function ready")
```

```python
# 9. CREATE UNIFIED CSVs (ALL GAMES COMBINED)

def create_unified_csvs(output_dir="nba_data"):
    """Combine all individual game CSVs into unified files"""
    
    if not os.path.exists(output_dir):
        print(f"‚ùå Directory {output_dir} does not exist")
        return
    
    # Find all CSV files
    csv_files = [f for f in os.listdir(output_dir) if f.endswith('.csv')]
    
    if not csv_files:
        print(f"‚ùå No CSV files found in {output_dir}")
        return
    
    print(f"Found {len(csv_files)} CSV files in {output_dir}")
    
    # Group files by data type
    file_groups = {}
    for file in csv_files:
        # Extract data type from filename (e.g., "line_scores_202310240LAL.csv")
        parts = file.split('_')
        if len(parts) >= 2:
            data_type = '_'.join(parts[:-1])  # Everything except the game_id
            if data_type not in file_groups:
                file_groups[data_type] = []
            file_groups[data_type].append(file)
    
    print(f"\nData types found: {list(file_groups.keys())}")
    
    # Combine each data type
    for data_type, files in file_groups.items():
        print(f"\nüìä Combining {len(files)} files for {data_type}...")
        
        all_data = []
        for file in files:
            filepath = os.path.join(output_dir, file)
            try:
                df = pd.read_csv(filepath)
                all_data.append(df)
                print(f"  ‚úì {file}: {len(df)} rows")
            except Exception as e:
                print(f"  ‚ùå Error reading {file}: {e}")
        
        if all_data:
            combined_df = pd.concat(all_data, ignore_index=True)
            combined_file = f"{output_dir}/UNIFIED_{data_type}.csv"
            combined_df.to_csv(combined_file, index=False, encoding='utf-8')
            
            print(f"‚úÖ Saved unified {data_type}: {len(combined_df)} rows ‚Üí {combined_file}")
            
            # Show summary
            print(f"   Shape: {combined_df.shape}")
            print(f"   Date range: {combined_df['game_date'].min()} to {combined_df['game_date'].max()}")
            
            if 'team' in combined_df.columns:
                print(f"   Unique teams: {combined_df['team'].nunique()}")
            
            if 'player' in combined_df.columns:
                print(f"   Unique players: {combined_df['player'].nunique()}")
    
    print("\nüéâ UNIFIED CSV CREATION COMPLETE!")

# Run unified CSV creation
create_unified_csvs()
```
