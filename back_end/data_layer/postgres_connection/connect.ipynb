```python
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy.ext.declarative import declarative_base
import os
from dotenv import load_dotenv


load_dotenv()

URL = os.getenv("DATABASE_URL")
print("Database URL:", URL)
```

```python
engine = create_engine(URL)
```

```python

local_session = sessionmaker(autocommit=False, autoflush=False, bind=engine)
Base = declarative_base()
```

```python
def get_db():
    db = local_session()
    try:
        yield db
    finally:
        db.close()
```

```python
try:
    with engine.connect() as connection:
        print("Connection to the database was successful!")
except Exception as e:
    print(f"An error occurred while connecting to database: {e}")
```

```python
class User:
    def __init__(self, name:str, password:str, email:str):
        self.user_name = name
        self.__password = password
        self._email = email
        self.role = 'Normal User'
        
    def getPassword(self)-> str:
        return self.__password
    
    def getRole(self) -> None:
        print('I am a', self.role)
        
```

```python
class Admin(User):
    def __init__(self, name, password, email):
        super().__init__(name, password, email )
        self.role = 'Admin User'
    def getRole(self):
        print('I am a ', self.role)
        
```

```python
class Owner(User):
    
    def __init__(self, name, password, email):
        super().__init__(name, password, email)
        self.role = 'Product Owner User'
    def getRole(self):
        print('I am a ', self.role)
        
```

```python
normal_user = User('Xolani','123','xolani@emai.co.za')
normal_user.getRole()
print(normal_user.user_name)
```

```python
admin_user = Admin('thabiso','admin_1','admin@emai.co.za')
admin_user.getRole()
print(admin_user.user_name)
```

```python
owner_user = Owner('small', 'owner_1', 'owner@email.co.za')
owner_user.getRole()
print(owner_user.getPassword())
print(owner_user.user_name)
```
